```python
import string
import emoji
import re
def preprocess_text(text):
  text = str(text)
  # remove url from text
  text = re.sub(r'http\S+', '', text)
  # remove the punctuations and emojis from the text
  text = "".join([c for c in text if c not in string.punctuation])
  text = "".join([c for c in text if c not in emoji.UNICODE_EMOJI])
  
  return text
```

```python
from stopwordsiso import stopwords
import jieba
def stop_word(text):
    stopword = stopwords(["zh"])
    text = jieba.cut(text)
    text = " ".join(text)
    text = " ".join([word for word in text.split() if word not in stopword])
    return text
```

```python
censored_df[1] = censored_df[1].apply(preprocess_text)
censored_df[1] = censored_df[1].apply(stop_word)
```

```python
from collections import Counter
def most_common_words_chinese(dataset):
  words = Counter()
  for text in dataset[1]:
    tokens = jieba.cut_for_search(str(text))
    for word in tokens:
      words[word] += 1
  most_common_words = words.most_common(50000)
  return most_common_words
```

```python
most_common_words = most_common_words_chinese(censored_df)                                                     
```

```python
import re
def remove_english_chars(most_common_words):
  word_list = []
  for word in most_common_words:
    if re.match(r'[a-zA-Z0-9 “”]', word[0]):
      pass
    else:
      word_list.append((word[0],word[1]))
  return word_list
```

```python
most_common_words = remove_english_chars(most_common_words)
```

```python
def make_word_list(words):
  word_list = {}
  for word in words:
    synonyms_list = synonyms.nearby(word)
    for synonym in synonyms_list[0]:
      if synonym not in words:
        word_list[word] = synonym
        break
    else:
       word_list[word] = word
  return word_list
```

```python
words_list = make_word_list(words_to_replace)
```

```python
words_list
```

```python
def replace_words(text):
  for word in text.split():
    if word in words_list:
      text = text.replace(word, words_list[word])
  return text
```

```python
censored_df[1] = censored_df[1].apply(replace_words)
```
